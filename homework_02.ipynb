### Задача 1: "Классический FizzBuzz"

```python

    int_input = int(input("Введите любое целое положительное число"))
    for i in range(1, int_input + 1): # int_input + 1 тк range берет невключительно по int_input
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print (i)
```

### Задача 2: "Интерактивный менеджер списка покупок"

```python
shopping_list = []

while True:
    print("Введите любую цифру из предложенного меню")
    print('"1. Добавить", "2. Показать", "3. Выйти"')
    usr_inpt = int(input("Введите цифру"))
    if usr_inpt == 1:
         print("Введите продукт для списка покупок")
         shp_inpt = input()
         shopping_list.append(shp_inpt)
    elif usr_inpt == 2:
        if len(shopping_list) == 0:
            print( "Список покупок пуст.")
        else: 
            for index, shop  in enumerate(shopping_list): #для красивого вывода списка покупок
            # По умолчанию нумерация с 0, добавим 1 для красоты
                print(f"{index + 1}. {shop}")
    elif usr_inpt == 3:
          print( '''Программа завершена. 
Ждем вас снова!''')
          break # Цикл прекращает работу
    else: 
        print("Неизвестная команда, попробуйте снова.")
```
